```python
import os 
from os.path import join 
from glob import glob 
```

```python
base_path = '/work/mflora/SummaryFiles'
dates = [d for d in os.listdir(base_path) if '.txt' not in d]

count = 0 
for d in dates:
    times = [t for t in os.listdir(join(base_path,d)) if 'basemap' not in t]
    for t in times:
        path = join(base_path,d,t)
        # Need times indices from 24-72
        files = glob(f'/work/mflora/SummaryFiles/{d}/{t}/wofs_ENS_[2-7]*')
        files.sort()
        #files[4:]
        if len(files)==53:
            count+=1 
        else:
            print(len(files))
```
